```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis=1)
```

```python
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC(gamma='auto')
```

```python
model_svc.fit(X_train, y_train)
```

```python
model_svc.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svc),{
    'C' : [1, 10, 20, 30],
    'kernel' : ['linear', 'rbf']
}, cv = 5, return_train_score=False )
```

```python
classifier.fit(X, y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svc),{
    'C' : [1, 10, 20, 30],
    'kernel' : ['linear', 'rbf']
},n_iter = 4, cv = 5, return_train_score=False )
```

```python
classifier_r.fit(X, y)
```

```python
results_r = pd.DataFrame(classifier_r.cv_results_)
```

```python
results_r
```

```python
results_r[['param_C', 'param_kernel', 'mean_test_score']]
```
